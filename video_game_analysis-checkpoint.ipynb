# Step 5. Test the following hypotheses

 ### Part 1.
We calculate t-test for the means of two **independent** samples of populations.  
We will use the golden standard of 5% as an alpha.<br/><br/><br/>
The **null** hypthosesis is that average user ratings of the Xbox One and PC platforms are the same. μ1 = μ2  
The **laternative** hypothesis is that they are different.

```python
xone_sample = games_data_new[games_data_new['platform'] == 'XOne']['user_score']
pc_sample = games_data_new[games_data_new['platform'] == 'PC']['user_score']
print('XOne variance is:',xone_sample.std() ** 2)
print('PC variance is:', pc_sample.std() ** 2)
```

```python
from scipy import stats as st



sample_1 = xone_sample
sample_2 = pc_sample

alpha = .05 # critical statistical significance level
                        # if the p-value is less than alpha, we reject the hypothesis

results = st.ttest_ind(
        sample_1, 
        sample_2,
equal_var = False)

print('p-value: ', results.pvalue)

if (results.pvalue < alpha):
        print("We reject the null hypothesis")
else:
        print("We can't reject the null hypothesis")
```

We **reject** the null hypothesis, therefore probably **alterantive** hypothesis is true.
_____
### Part 2.
The **null** hypthosesis is that average user ratings of the Action and Sports genre are the same. μ1 = μ2   
The **alternative** hypothesis is that they are different.

```python
action_sample = games_data_new[games_data_new['genre'] == 'Action']['user_score']
sport_sample = games_data_new[games_data_new['genre'] == 'Sports']['user_score']
print('Action variance is:',action_sample.std() ** 2)
print('Sport variance is:', sport_sample.std() ** 2)
```

```python
sample_11 = action_sample
sample_22 = sport_sample

alpha = .05 # critical statistical significance level
                        # if the p-value is less than alpha, we reject the hypothesis

results = st.ttest_ind(
        sample_11, 
        sample_22,
equal_var = False)

print('p-value: ', results.pvalue)

if (results.pvalue < alpha):
        print("We reject the null hypothesis")
else:
        print("We can't reject the null hypothesis")
```

**We can't reject the null hypothesis**  
Therefore based on the sample we may conclude that average user ratings of the Action and Sports genre are **the same**.
